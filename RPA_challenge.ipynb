```python
import rpa as r
```

```python
r.init(visual_automation=True)
r.url("https://www.rpachallenge.com/")
r.click("//a[contains(@href,'challenge.xlsx')]")
r.close()
```

```python
import pandas as pd
df = pd.read_excel('challenge.xlsx',dtype=str)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df =df.dropna()
df
```

```python
r.init(visual_automation=True)
r.url("https://www.rpachallenge.com/")
r.click("//button[contains(.,'Start')]")
```

```python
def paste(element_identifier, text_to_paste):
    r.click(element_identifier)
    r.clipboard(text_to_paste)
    r.keyboard('[cmd]v')
r.paste = paste
```

```python
for i in range(len(df.axes[0])):
    r.paste('//*[@ng-reflect-name="labelPhone"]', df["Phone Number"][i])

    r.type("//LABEL[text()='First Name']/following-sibling::INPUT", df["First Name"][i])
    r.type("//LABEL[text()='Last Name']/following-sibling::INPUT", df["Last Name"][i])
    r.type("//LABEL[text()='Company Name']/following-sibling::INPUT", df["Company Name"][i])
    r.type("//LABEL[text()='Role in Company']/following-sibling::INPUT", df["Role in Company"][i])
    r.type("//LABEL[text()='Address']/following-sibling::INPUT", df["Address"][i])
    r.type("//LABEL[text()='Email']/following-sibling::INPUT", df["Email"][i])
    r.type("//LABEL[text()='Phone Number']/following-sibling::INPUT", df["Phone Number"][i])
    r.click('//*[@value="Submit"]')
```
